```python
# Get the number of rows from the user
num_rows = int(input('Enter the number of rows\n'))

# Loop to create the top half of the diamond
for i in range(1, num_rows + 1):
    # Calculate stars and spaces for the current row
    stars = 2 * i - 1
    spaces = num_rows - i
    # Print spaces and stars to form the pattern
    print(' ' * spaces + '*' * stars)

# Loop to create the bottom half of the diamond
# It starts from (num_rows - 1) and goes down to 1, moving in reverse order.
for i in range(num_rows - 1, 0, -1):
    # Calculate stars and spaces for the current row
    stars = 2 * i - 1
    spaces = num_rows - i
    # Print spaces and stars to form the pattern
    print(' ' * spaces + '*' * stars)
```

# OPPE_1: Mock-8 (3)



## Approach

1. **Initialize Flags**:
   - Start by initializing flags for diagonal, scalar, and identity as `True`.
2. **Iterate through Elements**:
   - Use nested loops to iterate through each element of the matrix.
   1. **Check for Non-Diagonal Elements**:
        - Inside the loop, check if any element outside the main diagonal is non-zero. If found, set the `isDiagonal` flag to `False`.
   2. **Check for Scalar Matrix**:
        - Check if diagonal elements are equal. If not, set the `isScalar` flag to `False`.
   3. **Check for Identity Matrix**:
        - Check if diagonal elements are equal to 1. If not, set the `isIdentity` flag to `False`.
3. **Return Matrix Type**:
   - After the loops, check the flags. If `isDiagonal` is `False`, return 'non-diagonal'. If `isIdentity` is `True`, return 'identity'. Otherwise, if `isScalar` is `True`, return 'scalar'; otherwise, return 'diagonal'.

## Code

```python
def matrix_type(M):
    # Get the size of the square matrix
    n = len(M)

    # Initialize flags to True assuming the matrix is of a certain type
    isDiagonal, isScalar, isIdentity = True, True, True

    # Iterate through each element in the matrix
    for i in range(n):
        for j in range(n):
            # Check for non-diagonal elements (elements outside the main diagonal)
            if i != j and M[i][j] != 0:
                isDiagonal = False

            # Check for scalar matrix (diagonal elements are equal)
            if i == j and M[i][j] != M[0][0]:
                isScalar = False

            # Check for identity matrix (diagonal elements are equal to 1)
            if i == j and M[i][j] != 1:
                isIdentity = False

    # If not diagonal, it is non-diagonal
    if not isDiagonal:
        return 'non-diagonal'

    # If it is identity, return 'identity'
    if isIdentity:
        return 'identity'

    # If it is scalar, return 'scalar', otherwise, return 'diagonal'
    return 'scalar' if isScalar else 'diagonal'
```

```python
# Example usage:
matrix1 = [[1, 0, 0],  # diagonal
           [0, 2, 0],
           [0, 0, 3]]

matrix2 = [[2, 0, 0],  # scalar
           [0, 2, 0],
           [0, 0, 2]]

matrix3 = [[1, 0, 0],  # identity
           [0, 1, 0],
           [0, 0, 1]]

matrix4 = [[1, 2, 3],  # non-diagonal
           [4, 5, 6],
           [7, 8, 9]]
```

```python
print(matrix_type(matrix1))
print(matrix_type(matrix2))
print(matrix_type(matrix3))
print(matrix_type(matrix4))
```

# OPPE_1: Mock-14 (9)



## Approach

1. **Convert Input to List**:
   - Change the input string of coins into a list of integers.
2. **Initialize Boxes**:
   - Create a list `boxes` with zero coins for each of the five boxes.
3. **Distribute Coins**:
   - Distribute coins into boxes based on the index modulo 5.
4. **Find Max Box**:
   - Find the box index with the most coins using the `max` function and `index` method.
5. **Print Result**:
   - Display the box number with the highest coins (add 1 for a 1-based index).

## Code

ðŸš¨ Example input: 3972894910

```python
# Convert the input string of coins into a list of integers
coins = [int(x) for x in input()]

# Initialize a list representing five boxes with zero coins initially
boxes = [0 for _ in range(5)]

# Iterate through each coin in the sequence and distribute them into boxes
for i in range(len(coins)):
    # Distribute the current coin to the box determined by the current index modulo 5
    boxes[i % 5] += coins[i]

# Find the index of the box with the maximum number of coins
max_box_index = boxes.index(max(boxes))

# Print the box number (add 1 to convert from zero-based index to 1-based index)
print(max_box_index + 1)
```

# OPPE_1: Mock-10 (5)



## Approach

- **Problem:** Calculate total cost of transactions in a store
- **Input:** List `[]` of transactions, each represented by a dictionary `{}`
  - Transaction ID ('TID')
  - List of items, each with details (name, price, quantity)
- **Output:** Summary list `[]` of transactions, each represented by a dictionary `{}`
  - Transaction ID ('TID')
  - Corresponding cost of transaction

## Code


```python
def get_summary(listOfTransactions):
    # Initialize an empty list to store the summary information
    summary_list = []

    # Iterate through each transaction in the input list
    for transaction in listOfTransactions:
        # Initialize variables to calculate the total cost for the transaction
        total_cost = 0

        # Iterate through each item in the transaction
        for item in transaction['Items']:
            # Calculate the cost for each item and add it to the total cost
            item_cost = item['Price'] * item['Qty']
            total_cost += item_cost

        # Create a summary dictionary for the transaction
        summary_dict = {
            "Cost": total_cost,
            "TID": transaction['TID']
        }

        # Add the summary dictionary to the list
        summary_list.append(summary_dict)

    # Return the final list of summary information for each transaction
    return summary_list
```

```python
# Example usage:
print(
    get_summary([
        {
            'TID': 'Gurunath_8528',
            'Items': [
                {'Name': 'Notebook', 'Price': 50, 'Qty': 4},
                {'Name': 'Pencil', 'Price': 10, 'Qty': 1},
                {'Name': 'Eraser', 'Price': 15, 'Qty': 1},
                {'Name': 'File', 'Price': 80, 'Qty': 1}
            ]
        }
    ])
)
```

```python
def get_summary(listOfTransactions):
    # Use a list comprehension to create a list of dictionaries containing the summary information
    return [{
        "Cost": sum(
            [item['Price'] * item['Qty'] for item in trans['Items']]
        ),
        "TID": trans['TID']
    } for trans in listOfTransactions]
```
